导入数据

```python
#导入处理数据包
import numpy as np
import pandas as pd

#导入数据
#训练数据集
train = pd.read_csv('train.csv')
#测试数据集
test = pd.read_csv('test.csv')
print('训练数据集：',train.shape,' 测试数据集：',test.shape)
```

把训练数据和测试数据进行合并,以便同时对两个数据集进行清洗，

```python
full = pd.concat([train, test], ignore_index=True)
print('合并后的数据集：', full.shape)
```

Age存在缺失数据；Fare存在空值（不合实际）

查看每一列的数据类型，和数据总数


```python
full.info()
```

年龄（Age）、船票价格（Fare）、船舱号（Cabin）、登船港口（Embarked）里有缺失数据

清洗数据
处理缺失值
年龄（Age）和船票价格（Fare）是数值类型数据，用平均数来填充缺失值

```python
full['Age']=full['Age'].fillna(full['Age'].mean())
full['Fare']=full['Fare'].fillna(full['Fare'].mean())
```

登船港口（Embarked）是字符串数据，且仅缺失2条数据，将缺失值填充为最频繁出现的值

```python
#查看登船港口（Embarked）数据
full['Embarked'].head()
```

```python
#分类变量Embarked，查找最常见的类别
full['Embarked'].value_counts()
```

```python
full['Embarked']= full['Embarked'].fillna('S')
```

船舱号（Cabin）是字符串数据，缺失数据比较多，缺失值填充为“U”，表示未知（Unknow）

```python
full['Cabin']=full['Cabin'].fillna('U')

#查看最终缺失值处理情况
full.info()
```

乘客性别（Sex）：male对应数值1，female对应数值0，使用Series.map()函数

```python
full['Sex']=full['Sex'].map({'male':1,'female':0})
full['Sex'].head()
```

登船港口（Embarked）：S，C，Q，使用get_dummies进行one-hot编码。

```python
EmbarkedDf = pd.DataFrame()
EmbarkedDf = pd.get_dummies(full['Embarked'],prefix='Embarked')
EmbarkedDf.head()
```

添加编码产生的变量到数据集full，并把登船港口(Embarked)删掉

```python
full = pd.concat([full,EmbarkedDf],axis=1)
full=full.drop('Embarked',axis=1)
```

客舱等级（Pclass）:1=1等舱，2=2等舱，3=3等舱，使用get_dummies进行编码

```python
PclassDf = pd.DataFrame()
PclassDf=pd.get_dummies(full['Pclass'],prefix='Pclass')
PclassDf.head()
```

添加编码产生的变量到数据集full，并把客舱等级(Pclass)删掉。

```python
full=pd.concat([full,PclassDf],axis=1)
full=full.drop('Pclass',axis=1)
```

从姓名（Name）中提取头衔：乘客头衔每个名字当中都包含了具体的称谓或者说是头衔
例如：'Braund, Mr. Owen Harris'，逗号前面的是“名”，逗号后面是‘头衔. 姓’，将这部分信息提取出来后可以作为非常有用的一个新变量，用来帮助预测。

```python
#定义函数：从姓名中获取头衔
def getTitle(name):
    str1 = name.split(',')[1]
    str2 = str1.split('.')[0]
    #strip() 方法用于移除字符串头尾指定的字符（默认为空格）
    str3=str2.strip()
    return str3

TitleDf=pd.DataFrame()
full['Name']=full['Name'].map(getTitle)
full['Name'].head()
```

Officer	政府官员
Royalty	王室（皇室）
Mr	已婚男士
Mrs	已婚妇女
Miss	年轻未婚女子
Master	有技能的人/教师

```python
title_mapDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
full['Name']=full['Name'].map(title_mapDict)
TitleDf = pd.get_dummies(full['Name'])
TitleDf.head()
```

```python
full = pd.concat([full,TitleDf],axis=1)
full= full.drop('Name',axis=1)
```

从客舱号（Cabin）中提取客舱类别：客舱号的首字母是客舱的类别，提取首字母，并进行one-hot编码。

```python
CabinDf = pd.DataFrame()
full['Cabin'] = full['Cabin'].map(lambda a:a[0])
CabinDf = pd.get_dummies(full['Cabin'],prefix='Cabin')
CabinDf.head()
```

```python
full = pd.concat([full,CabinDf],axis=1)
full = full.drop('Cabin',axis =1)
```

建立家庭人数和家庭规模：家庭人数=同代直系亲属数（Parch）+不同代直系亲属数（SibSp）+乘客自己。

```python
FamilyDf=pd.DataFrame()
FamilyDf['FamilySize']= full['Parch']+full['SibSp']+1
```

小家庭Family_Small：家庭人数=1；
中等家庭Family_Medium: 2<=家庭人数<=4；
大家庭Family_Large: 家庭人数>=5。

```python
FamilyDf['Family_Small']=FamilyDf['FamilySize'].map(lambda a:1 if a == 1 else 0)
FamilyDf['Family_Medium']=FamilyDf['FamilySize'].map(lambda a:1 if 2 <= a <=4 else 0)
FamilyDf['Family_Large']=FamilyDf['FamilySize'].map(lambda a:1 if a >= 5 else 0)
FamilyDf.head()
```

```python
full = pd.concat([full,FamilyDf],axis=1)
full.head()
```

```python
#原始数据集：特征
source_X = full_X.loc[0:sourceRow-1,:]
#原始数据集：标签
source_y = full.loc[0:sourceRow-1,'Survived']
#预测数据集：特征
pred_X = full_X.loc[sourceRow:,:]
print('原始数据集：',source_X.shape)
print('预测数据集：',pred_X.shape)
```

```python
from sklearn.metrics import f1_score, precision_score, accuracy_score
from sklearn.metrics import recall_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
import xgboost as xgb
import lightgbm as lgb

# Model training and hyperparameter tuning
def train_model(X_train, y_train, X_val, y_val):
    # RandomForest 随机森林
    rf = RandomForestClassifier(random_state=42)    # 初始化随机森林分类器，并设定随机种子，以确保每次运行时的结果一致
    param_grid_rf = { # 设置随机森林分类器的参数网格，用于网格搜索 (GridSearchCV) 调优
        'n_estimators': [100, 200,300],   # 'n_estimators': 树的数量列表
        'max_depth': [5, 10, 15],   # 'max_depth': 树的最大深度列表
        'min_samples_split': [2, 5, 10], # 'min_samples_split': 分割节点时所需的最小样本数
        'class_weight': ['balanced'] # 'class_weight': 设置类的权重，这里设置为平衡以应对类别不平衡
    }
     # 使用网格搜索来寻找最佳参数组合， cv=5 表示5折交叉验证，scoring='accuracy' 表示以准确率为优化目标
    grid_rf = GridSearchCV(rf, param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
     # 将网格搜索应用于训练集，以找到随机森林的最佳超参数
    grid_rf.fit(X_train, y_train)
     # 获取基于网格搜索获得的最佳随机森林模型
    best_rf = grid_rf.best_estimator_

    # XGBoost 梯度提升
    # 初始化XGBoost分类器，并设定随机种子；禁用label encoder，；使用对数损失作为评估指标；
    xgb_clf = xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
    param_grid_xgb = { # 设置XGBoost分类器的参数网格
        'n_estimators': [100, 200,300],   # 'n_estimators': 树的数量列表
        'max_depth': [3, 6, 9], # 'max_depth': 树的最大深度列表
        'learning_rate': [0.01, 0.05, 0.1],  # 'learning_rate': 学习率
        'subsample': [0.6, 0.8, 1.0]  # 'subsample': 子采样率，用于控制每棵树的样本采样比例
    }
    # 使用网格搜索来寻找XGBoost的最佳参数组合
    grid_xgb = GridSearchCV(xgb_clf, param_grid_xgb, cv=5, scoring='accuracy', n_jobs=-1)
     # 将网格搜索应用于训练集，以找到XGBoost的最佳超参数
    grid_xgb.fit(X_train, y_train)
    # 获取基于网格搜索获得的最佳XGBoost模型
    best_xgb = grid_xgb.best_estimator_

    # LightGBM 基于梯度提升决策树
    # 初始化LightGBM分类器，并设定随机种子
    lgb_clf = lgb.LGBMClassifier(random_state=42)
    param_grid_lgb = {  # 设置LightGBM分类器的参数网格
        'n_estimators': [100, 200,300],   # 'n_estimators': 树的数量列表
        'max_depth': [5, 10, 15],# 'max_depth': 树的最大深度列表
        'feature_fraction': [0.6, 0.8, 1.0] # 'feature_fraction': 特征子集采样率，用于控制每棵树的特征采样比例
    }
    # 使用网格搜索来寻找LightGBM的最佳参数组合
    grid_lgb = GridSearchCV(lgb_clf, param_grid_lgb, cv=5, scoring='accuracy', n_jobs=-1)
     # 将网格搜索应用于训练集，以找到LightGBM的最佳超参数
    grid_lgb.fit(X_train, y_train)
    # 获取基于网格搜索获得的最佳LightGBM模型
    best_lgb = grid_lgb.best_estimator_

    # Voting Classifier 投票分类器 集成模型
    # 初始化一个投票分类器，集成了随机森林、XGBoost和LightGBM三个最佳模型
    voting_clf = VotingClassifier(estimators=[
        ('rf', best_rf), # estimators参数定义了每个基模型的名称和模型对象
        ('xgb', best_xgb),   
        ('lgb', best_lgb)
    ], voting='soft') # voting='soft' 表示使用软投票，即使用各模型的预测概率进行加权投票
     # 在训练集上训练投票分类器，以便融合三个最佳模型的预测结果
    voting_clf.fit(X_train, y_train) 
     # 返回训练好的投票分类器模型
    return voting_clf


# Train the model
model = train_model(train_X, train_y, test_X, test_y)

# Evaluation
y_pred = model.predict(test_X)
accuracy = accuracy_score(test_y, y_pred)
precision = precision_score(test_y, y_pred)
recall = recall_score(test_y, y_pred)
f1 = f1_score(test_y, y_pred)

print(f"Validation Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1 Score: {f1:.2f}")

# Predict on test set and save submission file
y_test_pred = model.predict(pred_X)

output = pd.DataFrame({'PassengerId':full.loc[sourceRow:,'PassengerId'], 'Survived': y_test_pred.astype(int)})
output.to_csv('submission.csv', index=False)
print("预测结果已保存到submission.csv")
```
